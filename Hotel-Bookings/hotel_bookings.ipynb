DATA ANALYSIS AND VISUALIZATION

```python
cancelled_perc = df["is_canceled"].value_counts(normalize = True).round(2) 
print(cancelled_perc)

plt.figure(figsize = (5,4))
plt.title("Reservation status")
plt.bar(["Not canceled", "Canceled"], df["is_canceled"].value_counts(), edgecolor = 'k', width = 0.7)
plt.show()


# The booking cancelation status here is that 62% of hotel bookings are not canceled but 37%  of hotel bookings are still canceled. 
# This cancellation % may not  seem like a huge number in comparison to 62% but 37% is still a high number.  
# The figure shows that the cancelation is more than half, which seems alarming.
```

```python
plt.figure(figsize = (8,4))
ax1 = sns.countplot(x = "hotel", hue = "is_canceled", data = df, palette = "Blues")
legend_labels,_ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title("Reservation status in different hotels", size = 20)
plt.xlabel("Hotel")
plt.ylabel("Number of reservation")
plt.legend(["not canceled", "canceled"])
plt.show()

# Comparing Resort hotels and city hotels we see that city hotels have the highest cancelation rates.
```

```python
resort_hotel = df[df["hotel"] == "Resort Hotel"]
resort_hotel["is_canceled"].value_counts(normalize = True).round(2)

# The resort hotel cancelation is 27 %
```

```python
city_hotel = df[df["hotel"] == "City Hotel"]
city_hotel["is_canceled"].value_counts(normalize = True).round(2)

# The city hotel cancelation is 41 %
```

```python
resort_hotel = resort_hotel.groupby("reservation_status_date")[["adr"]].mean()
city_hotel = city_hotel.groupby("reservation_status_date")[["adr"]].mean()
```

```python
plt.figure(figsize = (20,8))
plt.title("Average Daily Rate In City And Resort Hotel", fontsize = 30)
plt.plot(resort_hotel.index, resort_hotel["adr"], label = "Resort hotel")
plt.plot(city_hotel.index, city_hotel["adr"], label = "Resort hotel")
plt.legend(fontsize = 15)
plt.show()

# According to the figure, the average daily rate for city hotels is less than the resort hotels
# The spikes indicate the rise in prices for hotels on weekends or holidays.
# The price for city hotels is less than resort hotels 
```

```python
df["month"] = df["reservation_status_date"].dt.strftime('%B')

# df["month"] = df["reservation_status_date"].dt.month

plt.figure(figsize = (16,8))
ax1 = sns.countplot(x = "month", hue = "is_canceled", data = df, palette = "bright")
legend_labels,_ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title("Reservation status per month", size = 30)
plt.xlabel("Month")
plt.ylabel("Number of reservation")
plt.legend(["not canceled", "canceled"])
plt.show()

# In the graph we can see that the cancelation is higher when the reservation is high which does not give us a clear picture about the cancelation
```

```python
plt.figure(figsize=(14, 8))
plt.title("ADR Per Month", fontsize = 20)
adr_per_month = df[df["is_canceled"] == 1].groupby("month")[["adr"]].sum().reset_index()
sns.barplot(x = "month", y = "adr", data = adr_per_month, edgecolor = 'k', width = 0.5)
plt.show()

# The month with the higher "Average daily rate" has the highest cancelation.
# From the above graph January has the highest cancelation which is explained by the graph given below which has the highest adr per month.
```

```python
cancelled_data = df[df["is_canceled"] == 1]
top_10_country = cancelled_data["country"].value_counts()[:10]
plt.figure(figsize = (8,8))
plt.title("Top 10 countries with reservation cancellation", fontsize = 20)
plt.pie(top_10_country, autopct = "%.2f", labels = top_10_country.index)
plt.show()

# Portugal has the highest cancellation. Hotels need to be more aware and provide more facilities to people from Portugal.           
```

```python
cancelled_df_adr = cancelled_data.groupby("reservation_status_date")[["adr"]].mean().round(2)                
cancelled_df_adr.reset_index(inplace = True)
cancelled_df_adr.sort_values("reservation_status_date", inplace = True)
cancelled_df_adr
```

```python
not_cancelled_data = df[df["is_canceled"] == 0]
not_cancelled_df_adr = not_cancelled_data.groupby("reservation_status_date")[["adr"]].mean().round(2)                
not_cancelled_df_adr.reset_index(inplace = True)
not_cancelled_df_adr.sort_values("reservation_status_date", inplace = True)
not_cancelled_df_adr
```

```python
plt.figure(figsize = (20,6))
plt.title("Average Daily Rate", fontsize = 20)
plt.plot(not_cancelled_df_adr["reservation_status_date"], not_cancelled_df_adr["adr"], label = "not_cancelled") 
plt.plot(cancelled_df_adr["reservation_status_date"], cancelled_df_adr["adr"], label = "cancelled") 
plt.legend(fontsize = 15)
plt.show()

# The data here seems to be spread and messy to understand
```
